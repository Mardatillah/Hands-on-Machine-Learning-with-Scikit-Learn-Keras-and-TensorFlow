# Chapter 3: Classification
---
Pada chapter ini, dibahas tentang tugas klasifikasi dalam machine learning, yang berfokus pada bagaimana mengelompokkan data ke dalam kategori tertentu. Klasifikasi adalah salah satu masalah yang paling umum ditemui dalam machine learning, dan bab ini mengupas berbagai teknik serta evaluasi model klasifikasi.

### Konsep Klasifikasi
Klasifikasi adalah masalah di mana kita mencoba untuk memprediksi kategori atau kelas dari sebuah data baru berdasarkan data pelatihan yang telah diberi label. Contoh masalah klasifikasi termasuk:
*   Klasifikasi email menjadi spam atau non-spam.
*   Klasifikasi gambar sebagai anjing atau kucing.

### Langkah-langkah dalam Klasifikasi
1. Persiapan Data
    *   Seperti pada semua proyek machine learning, tahap pertama adalah menyiapkan data. Data harus dibersihkan dan diformat untuk memastikan bahwa algoritma dapat memprosesnya dengan baik.
2. Memilih Algoritma Klasifikasi
    *   Ada berbagai algoritma yang dapat digunakan untuk klasifikasi, seperti:
        *   **Logistic Regression:** Algoritma dasar untuk klasifikasi biner.
        *   **k-Nearest Neighbors (k-NN):** Algoritma berbasis instance yang mengklasifikasikan data berdasarkan kedekatannya dengan data lain.
        *   **Support Vector Machines (SVM):** Algoritma yang efektif untuk masalah klasifikasi yang lebih kompleks.
        *   **Decision Trees dan Random Forests:** Algoritma berbasis pohon keputusan yang dapat digunakan untuk klasifikasi.
3. Melatih Model
    *   Data pelatihan digunakan untuk melatih model klasifikasi yang dipilih. Model ini belajar dari contoh data untuk memahami pola yang dapat digunakan untuk memprediksi kelas data baru.
4. Evaluasi Model
    *   Setelah model dilatih, evaluasi dilakukan untuk menilai seberapa baik model dalam mengklasifikasikan data baru. Beberapa metrik evaluasi yang digunakan antara lain:
        *   **Akurasi:** Persentase prediksi yang benar.
        *   **Confusion Matrix:** Matriks yang menunjukkan jumlah prediksi benar dan salah untuk setiap kelas.
        *   **Precision, Recall, dan F1-Score:** Metrik yang lebih mendalam untuk mengevaluasi model, terutama dalam kasus ketidakseimbangan kelas.
        *   **ROC Curve dan AUC:** Digunakan untuk menilai model dalam konteks klasifikasi biner, dengan memplot trade-off antara true positive rate dan false positive rate.  

### Metode Evaluasi
*   **Confusion Matrix:** Digunakan untuk menggambarkan kinerja model pada data uji. Matriks ini menunjukkan jumlah prediksi benar dan salah untuk setiap kelas.
*   **Precision dan Recall:** Precision mengukur berapa banyak prediksi positif yang benar, sedangkan recall mengukur berapa banyak data positif yang berhasil diprediksi oleh model.
*   **Precision-Recall Trade-off:** Ketika meningkatkan precision, recall bisa menurun dan sebaliknya. Pemahaman trade-off ini penting dalam memilih model yang tepat.
*   **ROC Curve:** Menggambarkan kemampuan model untuk membedakan kelas positif dan negatif pada berbagai ambang batas.

### Multiclass Classification
*   Untuk masalah yang melibatkan lebih dari dua kelas (misalnya, mengklasifikasikan jenis buah), teknik seperti **One-vs-Rest (OvR)** atau **One-vs-One (OvO)** digunakan untuk menangani lebih dari dua kelas.

### Error Analysis
*   Melakukan analisis kesalahan adalah langkah penting dalam meningkatkan model klasifikasi. Dengan melihat prediksi yang salah, kita dapat menemukan pola yang belum dipahami oleh model dan membuat perbaikan.

### Multilabel Classification
*   Berbeda dengan klasifikasi standar, di mana setiap instance hanya dapat memiliki satu label, **multilabel classification** memungkinkan satu instance memiliki lebih dari satu label. Misalnya, sebuah artikel berita bisa termasuk dalam kategori "politics" dan "economy" sekaligus.

### Model Selection
*   Seleksi model merupakan langkah penting dalam proyek klasifikasi. Berbagai model diuji, dan yang terbaik dipilih berdasarkan evaluasi kinerja. Penggunaan teknik seperti cross-validation dapat membantu memberikan gambaran yang lebih realistis tentang kinerja model di dunia nyata.

### Poin-poin Utama:
*   Klasifikasi adalah tugas untuk memprediksi kategori dari data baru.
*   Algoritma umum untuk klasifikasi termasuk Logistic Regression, k-NN, SVM, Decision Trees, dan Random Forests.
*   Evaluasi model menggunakan metrik seperti akurasi, confusion matrix, precision, recall, F1-score, dan ROC curve.
*   Dalam multiclass classification, teknik seperti One-vs-Rest digunakan untuk menangani lebih dari dua kelas.
*   Multilabel classification memungkinkan data memiliki lebih dari satu label.
*   Error analysis membantu dalam memahami kesalahan model dan meningkatkan kinerjanya.
