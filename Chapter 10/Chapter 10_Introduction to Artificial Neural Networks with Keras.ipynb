# Chapter 10: Introduction to Artificial Neural Networks with Keras
---
Chapter ini memperkenalkan Artificial Neural Networks (ANNs), yang terinspirasi oleh cara kerja otak manusia, dan bagaimana membangun serta melatih jaringan saraf menggunakan Kerasâ€”sebuah API tingkat tinggi untuk TensorFlow.

1. From Biological to Artificial Neurons
    *   Neuron biologis adalah sel yang mentransmisikan informasi melalui sinapsis. Neuron buatan dalam Artificial Neural Networks (ANNs) meniru fungsi ini, di mana neuron menerima input, memprosesnya, dan memberikan output.
    *   Model Neuron dalam ANNs terdiri dari tiga komponen utama: input, fungsi aktivasi, dan output.

2. The Perceptron
    *   Perceptron adalah model neuron pertama yang ditemukan oleh Frank Rosenblatt pada 1950-an. Perceptron digunakan untuk klasifikasi biner dan memiliki satu lapisan dengan satu neuron.
        *   Model ini mengklasifikasikan input ke dalam dua kelas (misalnya, 0 atau 1), dan belajar dengan mengupdate bobot berdasarkan kesalahan pada prediksi.
        *   Meskipun sederhana, Perceptron hanya dapat menangani masalah linier dan tidak dapat menyelesaikan masalah yang lebih kompleks.

3. The Multilayer Perceptron (MLP) and Backpropagation
    *   Multilayer Perceptron (MLP) adalah perluasan dari Perceptron, yang terdiri dari beberapa lapisan neuron: lapisan input, lapisan tersembunyi (hidden layers), dan lapisan output.
    *   Backpropagation adalah algoritma yang digunakan untuk melatih MLP. Backpropagation menghitung kesalahan di lapisan output dan mendistribusikan kesalahan ini kembali ke lapisan-lapisan sebelumnya, memperbarui bobot melalui gradient descent untuk meminimalkan kesalahan.

4. Building an Image Classifier Using the Sequential API
    *   Sequential API dari Keras memungkinkan pembuatan dan pelatihan jaringan saraf dengan lapisan-lapisan yang disusun secara berurutan. Prosesnya mencakup:
        *   Menambahkan Convolutional Layers untuk ekstraksi fitur dari gambar.
        *   Menambahkan Pooling Layers untuk mengurangi dimensi data gambar.
        *   Menambahkan Dense Layers untuk klasifikasi akhir.
    *   Softmax Activation digunakan di lapisan output untuk mengklasifikasikan gambar ke dalam kelas yang sesuai.

5. Building a Regression MLP Using the Sequential API
    *   MLP untuk regresi digunakan untuk memprediksi nilai kontinu, seperti harga rumah atau suhu. Dalam hal ini, lapisan output menggunakan linear activation untuk menghasilkan output kontinu.

6. Building Complex Models Using the Functional API
    *   Functional API memungkinkan pembuatan model yang lebih fleksibel dan kompleks, seperti model dengan beberapa input, output, atau lapisan yang saling berbagi.
    *   Model ini memungkinkan koneksi antar lapisan yang lebih rumit dibandingkan dengan Sequential API.

7. Using the Subclassing API to Build Dynamic Models
    *   Subclassing API adalah cara paling fleksibel untuk membangun model dalam Keras. Pengguna dapat membuat kelas turunan dari tf.keras.Model untuk mendefinisikan arsitektur jaringan yang dinamis dan lebih kompleks, memberikan kontrol penuh terhadap proses pelatihan dan prediksi.

8. Saving and Restoring a Model
    *   Keras menyediakan cara untuk menyimpan model yang telah dilatih menggunakan metode model.save(). Model yang disimpan dapat dipulihkan kapan saja menggunakan tf.keras.models.load_model().

9. Using Callbacks
    *   Callbacks adalah fungsi yang dipanggil pada berbagai tahap selama pelatihan model. Keras menyediakan berbagai callback, seperti:
        *   EarlyStopping untuk menghentikan pelatihan saat tidak ada peningkatan pada model.
        *   ModelCheckpoint untuk menyimpan model pada titik terbaik selama pelatihan.

10. Using TensorBoard for Visualization
    *   TensorBoard adalah alat visualisasi yang digunakan untuk memantau pelatihan model, melihat grafik dari loss dan akurasi selama pelatihan, serta memvisualisasikan struktur model neural network.

11. Fine-Tuning Neural Network Hyperparameters
    *   Hyperparameter tuning adalah langkah penting dalam melatih jaringan saraf. Beberapa hyperparameter yang perlu disesuaikan adalah jumlah lapisan tersembunyi, jumlah neuron per lapisan, dan laju pembelajaran (learning rate).
        *   Teknik seperti grid search atau random search digunakan untuk menemukan kombinasi hyperparameter terbaik.

### Poin-poin Utama:
*   Perceptron adalah model neuron dasar, sementara MLP dengan backpropagation memungkinkan pelatihan jaringan saraf yang lebih dalam dan lebih kompleks.
*   Keras Sequential API mempermudah pembuatan model neural network berlapis, sementara Functional API dan Subclassing API menawarkan fleksibilitas untuk model yang lebih kompleks.
*   Callbacks dan TensorBoard membantu dalam memantau dan mengoptimalkan pelatihan model.
*   Hyperparameter tuning dan fine-tuning penting untuk meningkatkan kinerja jaringan saraf.
