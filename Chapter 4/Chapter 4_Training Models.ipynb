# Chapter 4: Training Models
---
Pada chapter ini, dibahas tentang cara melatih model machine learning dengan menggunakan berbagai teknik, termasuk regresi linear, optimasi dengan algoritma gradient descent, dan penerapan regularisasi untuk mencegah overfitting.

1. Linear Regression
    *   **Regresi Linear** adalah metode dasar dalam machine learning untuk memprediksi nilai kontinu (misalnya, harga rumah). Model ini mencari hubungan linier antara variabel independen (fitur) dan variabel dependen (target).
    *   **Normal Equation** digunakan untuk mencari parameter model tanpa perlu menggunakan algoritma optimisasi. Meskipun cepat untuk dataset kecil, metode ini tidak efisien untuk dataset besar.
2. Gradient Descent
    *   Gradient Descent adalah algoritma optimasi yang digunakan untuk menemukan parameter model (misalnya, bobot regresi) yang meminimalkan fungsi kerugian (loss function).
        *   **Batch Gradient Descent** menghitung gradien menggunakan seluruh dataset, sementara **Stochastic Gradient Descent (SGD)** memperbarui parameter untuk setiap data instance.
        *   **Mini-batch Gradient Descent** adalah kompromi antara keduanya, di mana data dibagi menjadi batch kecil untuk memperbarui parameter secara lebih efisien.
3. Regularized Linear Models
    *   Regularisasi adalah teknik untuk mencegah overfitting dengan menambahkan penalti pada parameter model. Dua jenis utama regularisasi adalah:
        *   **Ridge Regression (L2 Regularization):** Menambahkan penalti terhadap kuadrat bobot model.
        *   **Lasso Regression (L1 Regularization):** Menambahkan penalti terhadap nilai absolut bobot, yang dapat mengarah pada penghapusan fitur yang tidak penting.
        *   **Elastic Net:** Kombinasi antara Ridge dan Lasso, memberikan fleksibilitas dalam memilih regularisasi.
4. Learning Curves
    *   Learning Curves adalah alat yang digunakan untuk memvisualisasikan kinerja model terhadap data pelatihan dan data uji seiring dengan bertambahnya jumlah data pelatihan.
        *   Kurva ini membantu mengidentifikasi apakah model underfitting (kurang belajar) atau overfitting (terlalu cocok dengan data pelatihan).
5. Logistic Regression
    *   Logistic Regression digunakan untuk klasifikasi biner, memprediksi probabilitas kejadian dari dua kelas (misalnya, spam atau tidak spam). Meskipun bernama regresi, ini adalah algoritma klasifikasi.
        *   **Decision Boundaries:** Model ini memetakan input ke probabilitas yang kemudian diubah menjadi kategori menggunakan ambang batas.
6. Softmax Regression
    *   Softmax Regression adalah generalisasi dari Logistic Regression untuk masalah klasifikasi multi-kelas. Ini digunakan ketika ada lebih dari dua kelas yang harus diprediksi.

### Poin-poin Utama:
*   **Regresi Linear** digunakan untuk masalah regresi dan memodelkan hubungan linier antara fitur dan target.
*   **Gradient Descent** adalah algoritma optimasi untuk melatih model dan dapat diterapkan dengan batch, stochastic, atau mini-batch.
*   **Regularisasi** (Ridge, Lasso, Elastic Net) mencegah overfitting dengan menambahkan penalti pada bobot model.
*   **Learning Curves** membantu mendiagnosis underfitting atau overfitting.
*   **Logistic Regression** dan **Softmax Regression** digunakan untuk klasifikasi, dengan Softmax digunakan untuk masalah klasifikasi multi-kelas.
