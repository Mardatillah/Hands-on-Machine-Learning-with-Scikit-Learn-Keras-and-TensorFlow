# Chapter 9: Unsupervised Learning Techniques
---
Chapter ini membahas **Unsupervised Learning**, yaitu metode machine learning di mana model belajar dari data yang tidak memiliki label atau target. Tujuan utama unsupervised learning adalah menemukan pola atau struktur tersembunyi dalam data.

1. Clustering
    *   **Clustering** adalah teknik unsupervised learning yang digunakan untuk mengelompokkan data ke dalam grup atau cluster yang memiliki kemiripan tinggi satu sama lain.
        *   **K-Means** adalah algoritma clustering yang paling populer, di mana data dibagi menjadi **k** cluster berdasarkan kedekatannya dengan pusat cluster.
        *   **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)** adalah algoritma yang mengelompokkan data berdasarkan kerapatan dan dapat mengidentifikasi noise (data yang tidak termasuk dalam cluster).
        *   **Hierarchical Clustering** menghasilkan sebuah dendrogram yang menunjukkan hubungan hierarkis antar data dan memungkinkan pemisahan data pada berbagai tingkat kedalaman.
2. Limits of K-Means
    *   **K-Means** memiliki beberapa keterbatasan:
        *   Tidak dapat menangani data yang tidak berbentuk bola atau tidak terdistribusi secara seragam.
        *   K-Means mengharuskan jumlah cluster **k** ditentukan sebelumnya, yang kadang sulit diputuskan.
        *   Sensitif terhadap inisialisasi centroid awal, yang bisa menyebabkan hasil yang berbeda pada percobaan yang berbeda.

3. Using Clustering for Image Segmentation
    *   Clustering dapat digunakan untuk **image segmentation**, yaitu membagi gambar menjadi beberapa bagian berdasarkan kesamaan fitur (misalnya, warna atau tekstur). Hal ini berguna dalam pengolahan gambar dan visi komputer.

4. Using Clustering for Preprocessing
    *   Clustering dapat digunakan untuk **preprocessing** data sebelum melatih model supervised. Misalnya, data yang terklaster dapat diperlakukan sebagai fitur tambahan atau untuk mengelompokkan data dengan karakteristik yang serupa.

5. Using Clustering for Semi-Supervised Learning
    *   Clustering dapat digunakan dalam **semi-supervised learning**, di mana sebagian data diberi label dan sebagian besar lainnya tidak. Clustering membantu untuk memperkirakan label pada data yang tidak diberi label dengan menganggap data dalam cluster yang sama memiliki label yang serupa.

6. Gaussian Mixture Models (GMM)
    *   **Gaussian Mixture Models** adalah model probabilistik yang mengasumsikan bahwa data berasal dari beberapa distribusi Gaussian. GMM dapat digunakan untuk clustering, tetapi memberikan model probabilistik yang lebih fleksibel dibandingkan dengan K-Means, karena dapat mengakomodasi bentuk cluster yang lebih kompleks.

7. Anomaly Detection Using Gaussian Mixtures
    *   **Anomaly Detection** adalah teknik untuk mengidentifikasi data yang berbeda dari sebagian besar data lain. GMM dapat digunakan untuk mendeteksi anomali dengan mengidentifikasi data yang memiliki probabilitas rendah untuk berasal dari distribusi yang dipelajari.

8. Selecting the Number of Clusters
    *   Salah satu tantangan dalam clustering adalah memilih jumlah cluster yang tepat. Beberapa metode yang dapat digunakan untuk memilih jumlah cluster **k** antara lain:
        *   **Elbow Method:** Mencari titik dimana penurunan dalam **within-cluster sum of squares** mulai melambat.
        *   **Silhouette Score:** Mengukur seberapa mirip suatu titik dengan cluster-nya dibandingkan dengan cluster lain.
        *   **Gap Statistic:** Membandingkan hasil clustering dengan distribusi acak untuk memilih jumlah cluster optimal.

9. Bayesian Gaussian Mixture Models
    *   **Bayesian Gaussian Mixture Models (BGMM)** adalah variasi dari GMM yang menganggap jumlah komponen Gaussian tidak diketahui dan dihitung berdasarkan data. Teknik ini menggunakan pendekatan Bayesian untuk memperkirakan distribusi data.

10. Other Algorithms for Anomaly and Novelty Detection
    *   Selain GMM, terdapat algoritma lain untuk deteksi anomali dan novelty, seperti:
        *   **Isolation Forests:** Menggunakan pendekatan berbasis pohon keputusan untuk memisahkan data yang tidak biasa.
        *   **One-Class SVM:** SVM yang dilatih hanya dengan data dari satu kelas untuk mendeteksi anomali.

### Poin-poin Utama:
*   **Clustering** digunakan untuk mengelompokkan data yang tidak memiliki label berdasarkan kemiripannya.
*   **K-Means** adalah algoritma yang paling populer untuk clustering, namun memiliki keterbatasan seperti sensitivitas terhadap inisialisasi dan jumlah cluster yang harus ditentukan sebelumnya.
*   **DBSCAN** dan **Hierarchical Clustering** adalah alternatif yang lebih fleksibel.
*   **Gaussian Mixture Models (GMM)** menawarkan pendekatan probabilistik yang lebih fleksibel untuk clustering.
*   **Anomaly detection** menggunakan teknik seperti GMM dan **Isolation Forests** untuk mendeteksi data yang tidak biasa.
*   Menentukan jumlah cluster yang tepat adalah tantangan, dengan metode seperti **Elbow Method** dan **Silhouette Score** untuk membantu memilihnya.
